```python
class fila:
    def __init__(self,data = None, prox = None):
        self.data = data    
        self.prox = prox
        
class pilha:
    def __init__(self,data = None, ant = None):
        self.data = data    
        self.ant = ant     
```

### Fila (first-in first-out)

```python
class line_list: #Inicia a fila
    def __init__(self):
        self.head = None
        self.shape = 0
    
    def getno(self, ind): #Pega um nó através do indice 
        pointer = self.head
        for i in range(ind):
            if pointer:
                pointer = pointer.prox
            else:
                raise IndexError("Index fora do escopo")
        return pointer
        
    
    def line_add(self, x): #Adiciona um item seguinta a fila
        if self.head:
            pointer = self.head
            while(pointer.prox):
                pointer = pointer.prox
            pointer.prox = fila(x)
        else:
            self.head = fila(x)
        self.shape += 1
        
    def line_drop(self): # Remove o primeiro item da fila, sendo o primeiro a entrar e o primeiro a sair
            self.head = self.head.prox
            self.shape -= 1
        
    def __len__(self): #Depois de remover o primeiro, verifica o tamanho da fila 
        return self.shape   

list = line_list()#Instancia da class

def create_line_ride(): #Usa a class ride na fila
    created_ride = ride(random.choice(ride.gender_array),
        str(df_rides.user_birthdate[random.randint(0,287321)])[0:10],
        random.choice(ride.uf_array),
        str(df_rides.station_start[random.randint(0,287320)]).strip()
        )
    list.line_add(created_ride)
    
for x in range(6):
    create_line_ride()    

def finish_ride_line(): #Termina um passeio (ride) na fila
    list.getno(0).data.finish_ride(df_stations.station[random.randint(0,48)])
    list.line_drop()
    
for x in range(list.shape):    
    finish_ride_line()
    time.sleep(1)
```

### Pilha (last in, first out)

```python
class pile_list: #Inicia a fila com topo vazio
    def __init__(self):
        self.top = None
        self.shape = 0
 
    def getno(self, ind):
        pointer = self.top
        # Percorre a pilha até o índice desejado (índice negativo representa contar de trás para frente)
        for i in range(ind):
            if pointer:
                pointer = pointer.ant
            else:
                raise IndexError("Index fora de alcance")
        
        return pointer
        
    def pile_add(self, elemento): #Adiciona itens a pilha
        pointer = pilha(elemento)
        pointer.ant = self.top
        self.top = pointer
        self.shape += 1         
        
    def pile_drop(self): # Remove itens do topo da pilha
        self.top = self.top.ant
        self.shape -= 1
        
    def __len__(self):
        return self.shape   


    
pile = pile_list()
    
def create_pile_ride(): #Usa class ride na pilha
    created_ride = ride(random.choice(ride.gender_array),
        str(df_rides.user_birthdate[random.randint(0,287321)])[0:10],
        random.choice(ride.uf_array),
        str(df_rides.station_start[random.randint(0,287320)]).strip()
        )
    pile.pile_add(created_ride)
    
for x in range(6):
    create_pile_ride()    

print("---------")

def finish_pile_ride(): #Termina o passeio na pilha
    pile.getno(-1).data.finish_ride(df_stations.station[random.randint(0,48)])
    pile.pile_drop()

for x in range(pile.shape):
    finish_pile_ride()
```

### Bubblesort

O Bubblesort é um dos algoritmos de ordenação mais simples, mas também um dos menos eficientes.
Funciona comparando repetidamente pares de elementos adjacentes e trocando-os se estiverem na ordem errada.
Continua esse processo até que nenhuma troca seja necessária, o que indica que a lista está ordenada.
Sua complexidade de tempo no pior caso é O(n^2), onde n é o número de elementos na lista, tornando-o ineficiente para grandes conjuntos de dados.

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]  # Troca elementos
                swapped = True
        if not swapped:
            break
    return arr
```

### Mergesort

O Mergesort é um algoritmo de ordenação dividir para conquistar que é conhecido por sua eficiência e estabilidade.
Divide a lista em sublistas menores, ordena essas sublistas e, em seguida, combina-as para obter a lista ordenada final.
Possui uma complexidade de tempo média e no pior caso de O(n log n), tornando-o eficiente para grandes conjuntos de dados.
É um algoritmo estável, o que significa que a ordem relativa de elementos iguais é preservada.

```python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]

        merge_sort(left_half)
        merge_sort(right_half)

        i = j = k = 0

        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1

        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1

        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1

    return arr
```

### Heapsort

O Heapsort é um algoritmo de ordenação baseado em uma estrutura de dados chamada heap binário.
Transforma a lista em uma estrutura de heap, onde o elemento máximo está sempre na raiz.
Repetidamente remove o elemento máximo da raiz e o coloca no final da lista, diminuindo o tamanho do heap.
Possui uma complexidade de tempo média e no pior caso de O(n log n) e é in-place (não requer memória adicional além da lista original).
Não é estável, ou seja, não necessariamente preserva a ordem relativa de elementos iguais.

```python
def heapsort(arr):
    def heapify(arr, n, i):
        largest = i
        left = 2 * i + 1
        right = 2 * i + 2

        if left < n and arr[left] > arr[largest]:
            largest = left

        if right < n and arr[right] > arr[largest]:
            largest = right

        if largest != i:
            arr[i], arr[largest] = arr[largest], arr[i]
            heapify(arr, n, largest)

    n = len(arr)

    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)

    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        heapify(arr, i, 0)

    return arr
```

### Quicksort

O Quicksort é um algoritmo de ordenação dividir para conquistar que escolhe um elemento pivô, reorganiza a lista de forma que os elementos menores que o pivô estejam à esquerda e os elementos maiores estejam à direita, e em seguida, ordena recursivamente as sublistas à esquerda e à direita do pivô.
Possui uma complexidade de tempo média de O(n log n) e um pior caso raro de O(n^2).
Pode ser bastante rápido na prática e é frequentemente usado como algoritmo de ordenação padrão em linguagens de programação como o Python (na função sorted() ou método list.sort()).
Sua escolha de pivô pode afetar seu desempenho; portanto, existem variações, como o Quicksort de Mediana de Três, para melhorar o desempenho em casos específicos.

```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr

    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]

    return quick_sort(left) + middle + quick_sort(right)
```
